##1. Stacks

Stack works on the principle of Last-in, first-out **(LIFO)**. Also, the inbuilt functions in Python make the code short and simple. To add an item to the top of the list, i.e., to push an item, we use append() function and to pop out an element we use pop() function. These functions work quiet efficiently and fast in end operations.


2 operations:


1.   Push
2.   Pop

Both take place in O(1) time #constant time



image.png

```python
stack = ["Apple", "Orange", "banana"]


stack.append("Peach")
stack.append("Mango")
print(stack)
  
# Removes the last item
print(stack.pop())
  
print(stack)
  
# Removes the last item
print(stack.pop())
  
print(stack)
```

##2. Queue

Queue works on the principle of First-in, first-out **(FIFO)**. Below is list implementation of queue. We use pop(0) to remove the first item from a list.

2 operations:


1.   Enqueue - O(1)
2.   Dequeue - O(n-1) ~ O(n)



image.png

```python
queue = ["Apple", "Orange", "banana"]
queue.append("Peach")
queue.append("Mango")
print(queue)
  
# Removes the first item
print(queue.pop(0))
  
print(queue)
  
# Removes the first item
print(queue.pop(0))
  
print(queue)
```

###2.1 Using Dequeue

In case of stack, list implementation works fine and provides both append() and pop() in O(1) time. When we use deque implementation, we get same time complexity.

```python
from collections import deque

queue = deque(["Apple", "Orange", "banana"])
print(queue)

queue.append("Mango")
print(queue)

queue.append("Peach")
print(queue)

print(queue.pop())                 
print(queue.pop())                 

print(queue)
```

##3. DFS

DFS stands for Depth-First Search. It is a graph traversal algorithm that explores a graph by visiting as far as possible along each branch before backtracking. In other words, it starts at a root node and explores as far as possible along each branch before backtracking.

<u>Applications</u>:


1.   Finding connected components
2.   Topological sorting
3.   Path-Finding
4.   Cycle detection
5.   Game tree search




```python
from IPython.display import Image
Image(url='https://upload.wikimedia.org/wikipedia/commons/7/7f/Depth-First-Search.gif?20090326120256')
```

```python
'''
     Tree:-

       A
     /   \
    B     C
   / \     \
  D   E     F


'''


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': []
}

visited = set()

def dfs(start):
    stack = [start]
    while stack:
        node = stack.pop()
        print(node)
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                stack.append(neighbor)

dfs('A')
print(list(visited))
```

##4. BFS

BFS (Breadth-First Search) is another graph traversal algorithm that visits all the vertices of a graph in breadth-first order, i.e., it visits all the vertices at the same level before moving down to the next level. In other words, it explores all the vertices at distance 1 from the starting vertex, then all the vertices at distance 2, and so on

<u>Applications</u>:


1.   Shortest Paths
2.   MST (minimum Spanning Tree)
3. Web crawling - BFS can be used for web crawling, where the goal is to visit all the pages on a website.



```python
from IPython.display import Image
Image(url='https://upload.wikimedia.org/wikipedia/commons/4/46/Animated_BFS.gif?20070330182553')
```

```python
'''
     Tree:-

          A
        /   \
       B     C
      / \     \
     D   E     F
    /    / 
   G    H

'''

# Dictionary
graph = {
    'A': ['B', 'C'], # Parent node (key) with children nodes as the values
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G'],
    'E': ['H'],
    'F': [],
    'G': [],
    'H': []
}

visited = set()

def bfs(start):
    queue = [start] # Queues follow FIFO  first in, first out
    while queue:
        node = queue.pop(0)

        print(node)
        
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                queue.append(neighbor)

bfs('A') # Parsing A as the start node
print(visited)

# Whenever an element is popped out,  the queue has to move forward, filling up the spot that was popped out
# The algorithm goes level by level  this is the characteristic of breadth-first search

# In the end, all of the nodes should have been visited 


```
